```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Identify numeric columns
numeric_columns = credit_card_data.select_dtypes(include=[np.number]).columns

# Convert non-numeric values to NaN
credit_card_data[numeric_columns] = credit_card_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Calculate the mean of each numeric column
column_means = credit_card_data[numeric_columns].mean()

# Replace missing values in numeric columns with the mean
credit_card_data_imputed = credit_card_data.copy()
credit_card_data_imputed[numeric_columns] = credit_card_data_imputed[numeric_columns].fillna(column_means)

# Display the resulting dataframe
print(credit_card_data_imputed)
```

```python
credit_card_data_imputed.isnull().sum()
```

```python
#distribution of legit transaction & fraudelent transactions
credit_card_data['Class'].value_counts()
```

```python
# sepearting the data for the analysis
legit = credit_card_data[credit_card_data.Class==0.0]
fraud = credit_card_data[credit_card_data.Class==1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare these values for both transaction
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=3)
```

```python
new_dataset = pd.concat([legit_sample , fraud ] , axis = 0 )
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
 new_dataset.groupby('Class').mean()
```

splitting the data into features and targets

```python
X= new_dataset.drop(columns='Class' , axis = 1 )
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

split the data into training data and testing data

```python
X_train , X_test , Y_train ,  Y_test = train_test_split(X , Y ,test_size=0.2 ,stratify=Y ,random_state =2)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model Training

Logistic regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train , Y_train)
```

Model evaluation

Accuracy score

```python
#accuracy on training data
X_train_predection= model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predection , Y_train)
```

```python
print('Acurracy on Training data:',training_data_accuracy)
```

```python
#accuracy on test data
X_test_predection = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predection , Y_test)
```

```python
print('Accuracy on test data :' ,test_data_accuracy)
```
